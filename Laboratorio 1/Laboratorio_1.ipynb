```python
# Ejemplo de DataFrame con un Ã­ndice
data = {
    2006: [1, 2, 3, 4],
    2011: [5, 6, 7, 8]
}

index = ['Spain', 'Portugal', 'France', 'Germany']
pivedu = pd.DataFrame(data, index=index)

# Eliminar etiquetas especÃ­ficas del Ã­ndice (si existen)
etiquetas_a_eliminar = [
    'Euro area (13 countries)', 'Euro area (15 countries)', 'Euro area (17 countries)',
    'Euro area (18 countries)', 'European Union (25 countries)', 'European Union (27 countries)',
    'European Union (28 countries)'
]

pivedu = pivedu.drop(labels=etiquetas_a_eliminar, errors='ignore')

print(pivedu)
```

```python
totalSum = pivedu. sum(axis = 1).sort_values(ascending = False)
totalSum. plot(kind = 'bar', style = 'b', alpha = 0.4,
title = "Total Values for Country")
```
